Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

## Seaborn:
Seaborn is a powerful Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating a variety of informative and attractive statistical plots.

#### Histograms:
Use: Histograms are used to visualize the distribution of a single continuous variable. They group data into bins and display the frequency or density of data points in each bin.
Seaborn Function: sns.histplot()

#### Scatter Plots:
Use: Scatter plots are used to visualize the relationship between two continuous variables. Each data point is represented as a marker, allowing you to identify patterns, correlations, and outliers.
Seaborn Function: sns.scatterplot()

#### Bar Plots:
Use: Bar plots are suitable for comparing different categories or groups. They can display the frequency, count, or aggregated values of a categorical variable.
Seaborn Function: sns.barplot()

#### Box Plots:
Use: Box plots (box-and-whisker plots) provide a summary of the distribution of a dataset, including the median, quartiles, and potential outliers. They are useful for visualizing the spread and central tendency of data.
Seaborn Function: sns.boxplot()

#### Heatmaps:
Use: Heatmaps are used to visualize the correlation or relationships between multiple variables in a dataset. They are particularly useful for identifying patterns in large datasets.
Seaborn Function: sns.heatmap()

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
* Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.legend(title="Event")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
* Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
#Box plot for 'age' by 'pclass'
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

#Box plot for 'fare' by 'pclass'
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x="price", hue="cut", kde=True, palette="husl")
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.legend(title="Cut")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
iris = sns.load_dataset("iris")


sns.pairplot(iris, hue="species")
plt.suptitle("Pair Plot of Iris Dataset by Species") # to provide a title for the pair plot.
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Creating a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(flights_pivot, annot=True, fmt='d', cmap='coolwarm', linewidths=0.5)
plt.title("Flights Dataset Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
